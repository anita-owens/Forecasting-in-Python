# Forecasting sales with advertising effect in Python

How to measure the effectiveness of advertising on sales.

## Load packages and import data

```python
#Locate imported packages
import sys
sys.path

#import pandas and numpy
import pandas as pd
import numpy as np

# From matplotlib, import pyplot under the alias plt
from matplotlib import pyplot as plt

#Import library
from prophet import Prophet

from prophet.utilities import regressor_coefficients

# Reading an excel file using Python
import os
```

```python
# Give the location of your file -
file = ('datasets/addata.xlsx')
 
#Load spreadsheet
data = pd.ExcelFile(file)

## Load a sheet into a DataFrame by name
df = data.parse('diet')

#Check results
print(df.head())
```

```python
print(df.sort_values(by='y'))
```

```python
from datetime import datetime
#Create date sequence of 36 months so we have a proper time object as our index
Date = pd.date_range('2011-01-01', periods=36, freq = 'MS')

print(Date)

#Add date sequence I just created to to dataframe
df['Date'] = Date

#Check results
print(df['Date'].head())
```

## Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

# Plot sales and advertising on the same time-series
ax.plot(df['Date'], df["Sales"], color = 'blue')

# Set the x-axis label
ax.set_xlabel('Date')

# Set the y-axis label
ax.set_ylabel('Total Sales', color = 'blue')

ax2 = ax.twinx()
ax2.plot(df['Date'], df["Advertising"], color = 'red')
ax2.set_ylabel('Total Advertising', color = 'red')

# Show the figure
plt.show()
```

## Data Prep

```python
#Rename columns
df.rename(columns = {'Date': 'ds', 'Sales': 'y', 'Advertising':'advertising'},inplace = True)

#Check results of renaming
print(df.columns)
```

```python
#Drop Month column
df = df.drop('Month', axis = 1)
```

```python
#Test Train Split
import random

random.seed(10)

# train-test split
test_size = 12 # 12 months of data for test set
train_df = df[:-test_size]
test_df = df[-test_size:]


print(train_df.tail())
```

## Model Fit

```python
#We need to initiate a Prophet object
m = Prophet(seasonality_mode='multiplicative')
#m = Prophet()
m.add_regressor('advertising')

#Fit model
m.fit(df)
```

Prophet's default mode for seasonality is additive.

## How did Prophet handle the data

```python
#Create forecast using predict function - if no future dataframe is created then the predict function will simply predict values on the historical data that was fit in the model
#We just want to see how Prophet handeled seasonality
forecast = m.predict()

#Plot the historical forecast
fig_a_model = m.plot(forecast)
plt.show()
```

```python
#Components plot
fig = m.plot_components(forecast)
```

There are 3 components: Trend, Yearly & extra regressor. If you have a dataset that has at least two years of data, Prophet will automatically detect and isolate yearly seasonality components. 

Trend shows that demand has been increasing linearly since 2010.

The yearly trend is very wavy peaks & dips showing high peak demand in May with the lowest being March/April.

Extra regressor is multiplicative. This is the advertising variable that shows the magnitude that advertising increases sales. When advertising is at it's highest, sales tend to increase 20%+. There is a whopping 600% decrease in March.

```python
#More information about the regressor coefficients
regressor_coefficients(m)
```

The mean value of advertising is 28.53 (the center column). The coef column is the expected value of the coefficient. The expected impact on y based on a one unit increase (in this case in USD dollars) in advertising. The expected impact on sales is 0.007056 or a .7% increase for every sale higher than 28.53. Regressors are always modeled as linear relationships.Prophet requires all additional regressors to have defined values in the future forecast.

## Create Future Forecast

```python
#Make future dataframe 
#Period equals 365 days times 1 (1 year)
future = m.make_future_dataframe(periods=365 * 1)
future_forecast = m.predict(future)
```
