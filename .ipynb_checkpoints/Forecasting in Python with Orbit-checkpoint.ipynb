```python
#import pandas and numpy
import pandas as pd
import numpy as np

# From matplotlib, import pyplot under the alias plt
import matplotlib.pyplot as plt

# Reading an excel file using Python
import os

%matplotlib inline
import orbit
#from orbit.utils.dataset import load_iclaims
from orbit.models import DLT
from orbit.diagnostics.plot import plot_predicted_data
from orbit.eda import eda_plot
```

```python
# Give the location of your file -
file = ('datasets/Promotiondata.xlsx')
 
#Load spreadsheet
data = pd.ExcelFile(file)

## Load a sheet into a DataFrame by name
sales_df = data.parse('Sheet1')

#Check results
print(sales_df.head())
```

```python
#from datetime import datetime
#Create date sequence of 36 months so we have a proper time object as our index
date = pd.date_range('2011-01-01', periods=36, freq = 'MS')

print(date)

#Add date sequence I just created to to dataframe
sales_df['date'] = date

#Check results
print(sales_df['date'].head())
```

```python
#Rename columns
sales_df.rename(columns={'promotion?': 'promotion', 'Month of year':'month_of_year'}, inplace = True)

#Check results
print(sales_df.columns)

#Check data types
sales_df.info()
```

```python
#Encode promotion variable (replace yes/no with either 1/0)
sales_df['promo'] = [1 if x == 'yes' else 0 for x in sales_df['promotion']]
      

print(sales_df.head())
```

```python
#Drop original promotion column
sales_df.drop(columns=['promotion'],axis=1,inplace=True)

print(sales_df.head())
```

```python
#Subset and rename columns needed for modeling
df = sales_df[["date", "sales", "promo"]]

#Check results
print(df.head())
```

# Exploratory Data Analysis (EDA)

```python
#Print edaplot as a heatmap
_ = eda_plot.ts_heatmap(df = df, date_col = 'date', seasonal_interval=12, value_col='sales')
```

```python
var_list = ["date", "sales", "promo"]
_ = eda_plot.correlation_heatmap(df, var_list = var_list, 
                                 fig_width=10, fig_height=6)
```

```python
_ = eda_plot.dual_axis_ts_plot(df=df, var1='promo', var2='sales', date_col='date')
```

# Train test split

```python
import random

random.seed(10)

# train-test split
test_size = 12 # 12 months of data for test set
train_df = df[:-test_size]
test_df = df[-test_size:]

test_df.shape

print(train_df.tail())
```

```python
print(train_df)
```

```python
print(test_df)
```

```python
#Fit forecast model
dlt = DLT(
  response_col='sales', date_col='date',
    regressor_col=['promo'],
    seasonality=12,
)

dlt.fit(df=train_df)
```

```python
# outcomes data frame
predicted_df = dlt.predict(df=test_df)

print(predicted_df.head())
```

```python
#Plot the predicted data
from orbit.diagnostics.plot import plot_predicted_data
plot_predicted_data(
  training_actual_df=train_df, predicted_df=predicted_df,
  date_col=dlt.date_col, actual_col=dlt.response_col,
  test_actual_df=test_df
)
```

```python
dlt.get_regression_coefs()
```

```python
import numpy as np
from orbit.diagnostics.metrics import smape

# to reverse the log-transformation
def smape_adjusted(x, y):
    x = np.exp(x)
    y = np.exp(y)
    return smape(x, y)

naive_smape = smape_adjusted(predicted_df['prediction'].values, test_df['sales'].values)
#reg_smape = smape_adjusted(predicted_df_reg['prediction'].values, test_df['sales'].values)
#reg_adjust_smape = smape_adjusted(predicted_df_reg_adjust['prediction'].values, test_df['sales'].values)

#print('Naive Model: {:.3f}\nRegression Model: {:.3f}\nRefined Regression Model: {:.3f}'.format(
   # naive_smape #, reg_smape, reg_adjust_smape
#))
```
